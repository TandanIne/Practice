```python
# 1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?

def func_check_pos_neg(n):
    '''This function checks whther the given number is positive or negative or zero'''
    logging.info("This is he start of the function func_check_pos_neg")
    try :
        if(n > 0):
            logging.info("Function has successfully completed")
            return "Positive"
        elif(n < 0):
            logging.info("Function has successfully completed")
            return "Negative"
        else:
            logging.info("Function has successfully completed")
            return("Zero")
        
    except Exception as e:
        print("Error has happened , Please log file for more information")
        logging.error("Error has happened in the function ")
        logging.exception(str(e))

n= 7
print("Given number is %s"%func_check_pos_neg(n))
```

```python
#2.	Write a Python Program to Check if a Number is Odd or Even?

def odd_even(n):
    '''This function returns whether the given number is odd or even'''
    try:
        logging.debug("*"*20)
        logging.info("This is the start of the function odd_even")
    
        if(n%2 ==0):
            logging.info("Function has successfully completed")
            return "Even"
        else:
            logging.info("Function has successfully completed")
            return("Odd")
    except Exception as e:
        print("Error has Occurerd, please check the log message")
        logging.error("Error has occured in the function ")
        logging.exception(str(e))

n=7
print("Given Number is %s"%odd_even(n))
        
```

```python
#3  Write a Python Program to Check Leap Year?
def isleapyear(year):
    '''This function checks whether the given year is leap year or not'''
    try:
        logging.debug("*"*40)
        logging.info("This is the start of the function of isleapyear()")
        if(year % 4 ==0 and year %100 != 0 ):
            logging.info("Function has been completed successfully")
            return "Leap Year"
        elif( year % 400 == 0):
            logging.info("Function has been completed successfully")
            return "Leap Year"
        else:
            logging.info("Function has been completed successfully")
            return("not leap year")
    except Exception as e:
        print("Error has occured please check the log message")
        logging.error("Error has occured in this function")
        logging.exception(str(e))
year = 700
print("Given year is %s"%isleapyear(year))
```

```python
#4 Write a Python Program to Check Prime Number?
def isprime(n):
    ''' This function checks whether the given number is prime number or not'''
    try:
        logging.debug("*"*40)
        logging.info("This is the start of the function isprime()")
        count = 0
        for i in range(2,(n)):
            if(n%i ==0):
                count = 1
                break
        logging.info("Function has successfully completed")
        if(count ==1):            
            return("Not Prime")
        else:
            return("Prime")
    except Exception as e:
        print("Error has occured, please check the log file for more information")
        logging.error("Error has occured")
        logging.exception(str(e))
        
n= 12
print("Given number is %s"%isprime(n))
```

```python
#Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?
try:
    logging.debug("*"*50)
    logging.info("This is the start of the problem 5")
    for i in range(10000):
        st = isprime(i)
        
        if(st == "Prime"):
            print("%s is Prime"%i)
    logging.info("Program has successfully completed")
except Exception as e:
    print("Error has occured, please check the log message for more information")
    logging.error("Error has occured")
    logging.exception(str(e))
```
